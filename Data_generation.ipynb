```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
from numpy import *
```

```python
def generate_sample(root_url="/content/",file_name="data",file_output="samples",generate_step_len=10,sigement_size=300,start=0,end=0):

  df=pd.read_csv(root_url+file_name+".csv") 

  samplefile = open(root_url+file_output+"_S"+".csv","a+", encoding='utf-8', newline='')
  writer = csv.writer(samplefile)
  data_len=len(df['index'])

  list0=list(np.arange(0,sigement_size,1))
  list1=[]
  list2=[]
  list3=[]
  list4=[]
  list5=[]
  list6=[]

  reference_resistance=100

  for step in range(data_len):
    list1.append(df['voltage'][step])
    list2.append(df['current'][step])
    list3.append(df['capacity'][step])
    list4.append(df['energy'][step])
    list5.append(df['total_capacity'][step])
    list6.append(df['total_energy'][step])
    if len(list1)>=sigement_size:
      if ((step>=start and step+sigement_size<=end) or end==0) and step%generate_step_len==0:
        columns=list0+list1+list2+list3+list4+list5+list6
        if df['label'][step] == 0:
          columns.append(np.log(reference_resistance/1e6))
        else:
          columns.append(np.log(reference_resistance/df['label'][step]))
        writer.writerow(columns)
        # del(list0[0])
      del(list1[0])
      del(list2[0])
      del(list3[0])
      del(list4[0])
      del(list5[0])
      del(list6[0])

  samplefile.close()
```
